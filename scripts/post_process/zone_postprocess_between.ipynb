This script loads all results from random runs of the between-county models (core district and sub-district). It computes averages for all accuracy metrics (precision, recall, F1). Also computes average for feature importances across all model runs.

```python
import os

import pandas as pd
```

```python
# Set test the the model you want to computing average results for
# test = "between_sub_"
test = "between_core_"
```

```python
file_path = os.path.join(os.getcwd(), "results")
```

```python
if test == "between_sub_":
    col_names = ["Residential High", "Residential Medium-High", "Residential Medium", "Residential Medium-Low", "Residential Low", "Agriculture", "Commercial", "Industrial", "Office Institution", "Open Space", "Mixed Use", "Planned Development", "Downtown"]
else:
    col_names = ["Residential", "Non-Residential", "Mixed-Use"]
```

```python
def process_report(file, col_names):
    class_report = pd.read_pickle(os.path.join(file_path, file, "class_report.pkl"))
    class_report = class_report.round(4)
    class_report.index = col_names + ['accuracy', 'macro average', 'weighted average', 'r2']
    return(class_report)
```

```python
r1 = process_report((test + "r1"), col_names)
r2 = process_report((test + "r2"), col_names)
r3 = process_report((test + "r3"), col_names)
r4 = process_report((test + "r4"), col_names)
r5 = process_report((test + "r5"), col_names)
r6 = process_report((test + "r6"), col_names)
r7 = process_report((test + "r7"), col_names)
r8 = process_report((test + "r8"), col_names)
r9 = process_report((test + "r9"), col_names)
r10 = process_report((test + "r10"), col_names)
r11 = process_report((test + "r11"), col_names)
r12 = process_report((test + "r12"), col_names)
r13 = process_report((test + "r13"), col_names)
r14 = process_report((test + "r14"), col_names)
r15 = process_report((test + "r15"), col_names)
```

```python
precision = pd.concat([r1.iloc[:,0], r2.iloc[:,0], r3.iloc[:,0], r4.iloc[:,0], r5.iloc[:,0],r6.iloc[:,0], r7.iloc[:,0], r8.iloc[:,0], r9.iloc[:,0], r10.iloc[:,0], r11.iloc[:,0], r12.iloc[:,0], r13.iloc[:,0], r14.iloc[:,0], r15.iloc[:,0]], axis=1)
precision_mean = precision.mean(numeric_only=True, axis=1)
precision_sd = precision.std(numeric_only=True, axis=1)
precision = pd.concat([precision, precision_mean, precision_sd], axis=1)
precision.columns = ["run 1", "run 2","run 3","run 4", "run 5", "run 6", "run 7","run 8","run 9", "run 10", "run 11", "run 12", "run 13", "run 14", "run 15", "mean", "sd"]
file_name = test + "precision_results.csv"
precision.to_csv(os.path.join(file_path, file_name))
precision
```

```python
recall = pd.concat([r1.iloc[:,1], r2.iloc[:,1], r3.iloc[:,1], r4.iloc[:,1], r5.iloc[:,1],r6.iloc[:,1], r7.iloc[:,1], r8.iloc[:,1], r9.iloc[:,1], r10.iloc[:,1], r11.iloc[:,1], r12.iloc[:,1], r13.iloc[:,1], r14.iloc[:,1], r15.iloc[:,1]], axis=1)
recall_mean = recall.mean(numeric_only=True, axis=1)
recall_sd = recall.std(numeric_only=True, axis=1)
recall = pd.concat([recall, recall_mean, recall_sd], axis=1)
recall.columns = ["run 1", "run 2","run 3","run 4", "run 5", "run 6", "run 7","run 8","run 9", "run 10", "run 11", "run 12", "run 13", "run 14", "run 15",  "mean", "sd"]
file_name = test + "recall_results.csv"

recall.to_csv(os.path.join(file_path, file_name))
recall
```

```python
f1 = pd.concat([r1.iloc[:,2], r2.iloc[:,2], r3.iloc[:,2], r4.iloc[:,2], r5.iloc[:,2],r6.iloc[:,2], r7.iloc[:,2], r8.iloc[:,2], r9.iloc[:,2], r10.iloc[:,2], r11.iloc[:,2], r12.iloc[:,2], r13.iloc[:,2], r14.iloc[:,2], r15.iloc[:,2]], axis=1)
f1_mean = f1.mean(numeric_only=True, axis=1)
f1_sd = f1.std(numeric_only=True, axis=1)
f1 = pd.concat([f1, f1_mean, f1_sd], axis=1)
f1.columns = ["run 1", "run 2","run 3","run 4", "run 5", "run 6", "run 7","run 8","run 9", "run 10", "run 11", "run 12", "run 13", "run 14", "run 15",  "mean", "sd"]
file_name = test + "f1_results.csv"
f1.to_csv(os.path.join(file_path, file_name))
f1
```

```python
support = pd.concat([r1.iloc[:,3], r2.iloc[:,3], r3.iloc[:,3], r4.iloc[:,3], r5.iloc[:,3],r6.iloc[:,3], r7.iloc[:,3], r8.iloc[:,3], r9.iloc[:,3], r10.iloc[:,3], r11.iloc[:,3], r12.iloc[:,3], r13.iloc[:,3], r14.iloc[:,3], r15.iloc[:,3]], axis=1)

proportion = support.iloc[0:len(col_names), 0:15].copy()
for col in range(0,15):
    for row in range(0,len(col_names)):
        proportion.iloc[row,col] = support.iloc[row,col]/sum(support.iloc[0:len(col_names),col])

proportion_mean = proportion.mean(numeric_only=True, axis=1)
proportion_sd = proportion.std(numeric_only=True, axis=1)
proportion = pd.concat([proportion, proportion_mean, proportion_sd], axis=1)
proportion.columns = ["run 1", "run 2","run 3","run 4", "run 5", "run 6", "run 7","run 8","run 9", "run 10", "run 11", "run 12", "run 13", "run 14", "run 15", "mean", "sd"]

file_name = test + "proportion_results.csv"
proportion.to_csv(os.path.join(file_path, file_name))

proportion
```

```python
results = pd.concat([precision.iloc[:,15:17], recall.iloc[:,15:17], f1.iloc[:,15:17], proportion.iloc[:,15:17]], axis = 1)
header = [["Precision", "Precision", "Recall", "Recall", "F1-score", "F1-score", "Proportion", "Proportion"],
          ["mean", "sd","mean", "sd","mean", "sd","mean", "sd"]]
results.columns = header
file_name = test + "averaged_results.csv"
results.to_csv(os.path.join(file_path, file_name))
results
```

```python
# get features from pred folder
pred = os.listdir(os.path.join(os.getcwd(), "pred"))
features = [f.replace('.tif', '') for f in pred]
# sort features alphabetically 
features = sorted(features,  key=lambda s: s.lower())
```

```python
def feature_importance(test):
    file = os.path.join(file_path, test, "feature_import.csv")
    fi = pd.read_csv(file, usecols=[1])
    fi = fi.rename(columns={'0': test})
    fi.index = features 
    return(fi)
```

```python
r1_fi = feature_importance(test + "r1")
r2_fi = feature_importance(test + "r2")
r3_fi = feature_importance(test + "r3")
r4_fi = feature_importance(test + "r4")
r5_fi = feature_importance(test + "r5")
r6_fi = feature_importance(test + "r6")
r7_fi = feature_importance(test + "r7")
r8_fi = feature_importance(test + "r8")
r9_fi = feature_importance(test + "r9")
r10_fi = feature_importance(test + "r10")
r11_fi = feature_importance(test + "r11")
r12_fi = feature_importance(test + "r12")
r13_fi = feature_importance(test + "r13")
r14_fi = feature_importance(test + "r14")
r15_fi = feature_importance(test + "r15")
```

```python
fi_concat = pd.concat([r1_fi, r2_fi, r3_fi, r4_fi, r5_fi,r6_fi, r7_fi, r8_fi, r9_fi, r10_fi, r11_fi, r12_fi, r13_fi, r14_fi, r15_fi], axis=1)
fi_concat['mean'] = fi_concat.mean(axis=1)
fi_concat['std'] = fi_concat.std(axis=1)
fi_concat
```

```python
file_name = test + "feature_importance.csv"
fi_concat.to_csv(os.path.join(file_path, file_name))
```
